```python

import random
import string
import pandas as pd
import numpy as np
import datetime
import os
import sys
```

```python

def load_csv(file):
    data = pd.read_csv(file, sep=';', encoding='utf-8')
    return data


def transform_to_md(data):
    md = data.to_markdown()
    return md


def save_to_md(data, file):
    with open(file, 'w') as f:
        f.write(data)
    print('Data saved to', file)

data = load_csv('faqs.csv')
md = transform_to_md(data)
save_to_md(md, 'faqs.md')
```
